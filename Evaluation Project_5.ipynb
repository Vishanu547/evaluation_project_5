```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
data = pd.read_csv('https://raw.githubusercontent.com/dsrscientist/dataset1/master/census_income.csv')
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
lab_enc = LabelEncoder()
```

```python
data1 = lab_enc.fit_transform(data['Workclass'])
data2 = lab_enc.fit_transform(data['Education'])
data3 = lab_enc.fit_transform(data['Marital_status'])
data4 = lab_enc.fit_transform(data['Occupation'])
data5 = lab_enc.fit_transform(data['Relationship'])
data6 = lab_enc.fit_transform(data['Race'])
data7 = lab_enc.fit_transform(data['Sex'])
data8 = lab_enc.fit_transform(data['Native_country'])
data9 = lab_enc.fit_transform(data['Income'])
```

```python
data['Workclass'] = data1
data['Education'] = data2
data['Marital_status'] = data3
data['Occupation'] = data4
data['Relationship'] = data5
data['Race'] = data6
data['Sex'] = data7
data['Native_country'] = data8
data['Income'] = data9
```

```python
plt.figure(figsize = [20,15], facecolor = 'g')
plotno = 1


for column in data:
    if plotno<=15:
        ax = plt.subplot(3,5,plotno)
        sns.distplot(data[column])
        plt.xlabel(column,fontsize=12)
        
    plotno+=1
    
    
plt.show()    
```

```python
x = data.drop(columns = ['Income'])
y = data.Income
```

```python
plt.figure(figsize = [20,15], facecolor = 'g')
plotno = 1


for column in x:
    if plotno<=15:
        ax = plt.subplot(3,5,plotno)
        plt.scatter(x[column],y)
        plt.xlabel(column,fontsize=12)
        plt.ylabel('AveragePrice',fontsize = 12)
                           
        
    plotno+=1
    
    
plt.show()    
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y,test_size = 0.25,random_state=8)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr = LogisticRegression()
```

```python
lr.fit(x_train,y_train)
```

```python
y_pred = lr.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(classification_report(y_test,y_pred))
```

```python
print(confusion_matrix(y_test,y_pred))
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn.fit(x_train,y_train)
```

```python
print(confusion_matrix(y_test,y_pred))
```
